```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
```

```python
def read_file(location):
    with open(os.path.expanduser(location), "r") as file:
        return pd.read_csv(file, sep=",")
```

```python
df = read_file("/Users/guillermocomesanacimadevila/Desktop/BH_analysis/dacid/co2-emissions-and-gdp.csv")
```

```python
df
```

```python
df2 = df[((df["Entity"] == "European Union (27)") | (df["Entity"] == "United States")) & (df["Year"].between(2010, 2020))]
```

```python
df2
```

```python
df2.shape
```

```python
df_eu = df2[df2["Entity"] == "European Union (27)"]
df_usa = df2[df2["Entity"] == "United States"]
```

```python
df_eu.shape, df_usa.shape # ensure shape = "equal"
```

```python
df_eu
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df_eu["Year"], df_eu["Annual CO₂ emissions"], 
         label="EU", marker="o", linestyle="-", linewidth=2, color="#228B22")

plt.plot(df_usa["Year"], df_usa["Annual CO₂ emissions"], 
         label="USA", marker="s", linestyle="--", linewidth=2, color="#317873")

plt.xlabel("Year", fontsize=14, labelpad=10, fontweight="bold")
plt.ylabel("Annual CO₂ emissions", fontsize=14, labelpad=10, fontweight="bold")
plt.title("Annual CO₂ emissions", fontsize=18, pad=15, fontweight="bold")
plt.xticks(df_eu["Year"], fontsize=12)
plt.gca().yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))
plt.grid(visible=True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12, loc="upper left", bbox_to_anchor=(1, 1), frameon=True, shadow=True, fancybox=True, borderpad=1)

plt.tight_layout()

plt.savefig("EU_vs_USA_CO2_fixed.png", dpi=300, bbox_inches="tight")
plt.show()
```
