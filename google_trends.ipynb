```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_file(location):
    with open(os.path.expanduser(location), "r") as file:
        return pd.read_csv(file, sep=",")
```

```python
df = read_file("~/Desktop/BH_analysis/Cleaned_and_Merged_Data.csv")
```

```python
df
```

```python
df.columns
```

```python
df.columns = ["NA", "Time", "Eco-Friendly Searches", "Probiotic Searches"]
```

```python
df
```

```python
plt.scatter(df["Time"], df["Eco-Friendly Searches"], label = "eco-friendly", marker = "o")
plt.scatter(df["Time"], df["Probiotic Searches"], label = "probiotics", marker = "o")

plt.xlabel("Time")
plt.ylabel("Number of Searches")
plt.title("Trends in Eco-Friendly and Probiotic Usage")
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(df["Time"], df["Eco-Friendly Searches"], label="Eco-Friendly Searches", marker="o", linestyle='-', linewidth=2)
plt.plot(df["Time"], df["Probiotic Searches"], label="Probiotic Searches", marker="s", linestyle='--', linewidth=2)

plt.xlabel("Time", fontsize=14, labelpad=10)
plt.ylabel("Number of Searches", fontsize=14, labelpad=10)
plt.title("Trends in Eco-Friendly and Probiotic Searches", fontsize=16, pad=15)
plt.legend(fontsize=12, loc='upper left', frameon=True, shadow=True)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.grid(visible=True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
